```python
from  sklearn.linear_model import LinearRegression
import numpy as np
import seaborn as sns
from tensorflow import keras
import matplotlib.pyplot as plt 
```

```python
import pandas as pd
data = pd.read_csv('D:/MLwork/.venv/ML_LAB/kc_house_data.csv')
data.head()
```

```python
data[['sqft_living']]= data[['sqft_living']].astype(int)
Y = data[['price']]
X = data[['sqft_living']]


```

```python
model = LinearRegression().fit(X,Y)
# y_pred = model.predict()
# y_pred
X_pred = np.array([[2570]])
y_pred = model.predict(np.array(X_pred))
y_pred
```

```python
plt.plot(X, model.predict(X), color='red', linewidth=2, label='Regression Line')
plt.scatter(X_pred, y_pred, color='yellow', marker='x', s=100, label='Predicted Points')
# plt.style.use('dark_background')

# plt.scatter(X, Y, color='blue', label='Actual Data')
plt.scatter(X_pred, y_pred, color='lime', marker='x', s=100, label='Predicted Points')

plt.show()
```
